# Visualizing The Data

```python
sns.lineplot(x='date_time',y='temperature',data=Weather)
```

```python
sns.barplot(Weather['wind_speed'],Weather['humidity'])

```

```python
sns.displot(Weather['humidity'])
```

```python
sns.jointplot(Weather['humidity'],Weather['wind_speed'])
```

```python
sns.jointplot(Weather['humidity'],Weather['wind_speed'],kind="hex")
```

```python
sns.jointplot(Weather['humidity'],Weather['wind_speed'],kind="kde")
```

```python
sns.stripplot(Weather['weather_type'],Weather['wind_speed'])
```

```python
sns.stripplot(Weather['weather_type'],Weather['wind_speed'],jitter = True)
```

```python
sns.boxplot(Weather['dew_point'],Weather['temperature'],hue=Weather['weather_type'])
```

```python
sns.barplot(Weather['dew_point'],Weather['wind_speed'],hue=Weather['weather_type'])
plt.xticks(rotation=75)
```

```python
sns.countplot(Weather['weather_type'])
```

```python
sns.pointplot(Weather['humidity'],Weather['temperature'],hue=Weather['weather_type'])
```

```python
sns.lmplot(x="humidity",y="temperature",data=Weather)
```

```python
sns.lmplot(x="humidity",y="temperature",hue="weather_type",data=Weather)
```
